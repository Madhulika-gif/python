### if statement
The if statement in Python is used to test a condition and execute code only if the condition is True.

```python
# syntax
# if  condition :
  # Body

mark = 34
if mark >= 35:  # 10 >= 35 False
    print("Pass")
else:
  print('Fail')
```

```python
# 1. check if a number is positive:

num = 10
if num > 0: # num greated than 0
    print("The number is positive")
```

```python
# 2. Check if a number is even

num = 4
if num % 2 == 0:
    print("Even number")
```

```python
# 3. Check if list is not empty
items = [1,2,3]
if items:
    print("List is not empty")
```

```python
# 4. Check string length
name = "Python"
if len(name) > 5:
    print("Long name")
```

```python
# 5. Check if character is a vowel
ch = 'a'
if ch in 'aeiou':
    print("Vowel")
```

### if - else statement:

An if...else statement in Python is a conditional control structure that allows your program to execute one block of code if a condition is True, and a different block if the condition is False.

```python
# 1. even or odd number:

num = 3
if num % 2 == 0:
    print("Even")
else:
    print("Odd")
```

```python
# 2. Check Positive or Negative
x = -5
if x >= 0:
    print("Positive")
else:
    print("Negative")
```

```python
# 3. String Length Check
name = "AI"
if len(name) > 3:
    print("Long name")
else:
    print("Short name")
```

```python
# 4. Check Voting Eligibility
age = 16
if age >= 18:
    print("Eligible to vote")
else:
    print("Not eligible to vote")
```

```python
# 5. Check Password Match:
password = "admin123"
confirm = "admin123"
if password == confirm:
    print("Password matched")
else:
    print("Password mismatch")
```

```python
print(10 % 2) # -- remainder
print(10 / 2) # -- quotient
```

### if - elif statement:
An if...elif statement in Python is used to check multiple conditions one by one. If the first condition is false, Python checks the next one using elif (short for else if). You can add as many elif blocks as needed.

```python
# 1. example: Grading System:
marks = 55

if marks >= 90:  # 90 above
    print("Grade A")
elif marks >= 75: # 75 above
    print("Grade B")
elif marks >= 60:   # 60 above
    print("Grade C")
else:
    print("Grade D")
```

```python
# example : 2 BMI Category:
bmi = 22.5

if bmi < 18.5:
    print("Underweight")
elif bmi < 25:
    print("Normal weight")
elif bmi < 30:
    print("Overweight")
else:
    print("Obese")
```

```python
# elif can be used to check alternate conditions
country=input("Name of the country:")
if country==country.upper():
    print("The font entered is ALL CAPITAL")
elif country==country.lower():
    print("The font is all lowercase.")
elif country==country.capitalize():
    print("The font received is appropriate.")
else:
    print("Mixed case received")



print("One of the 4 blocks of code was executed.")
```

### Logical Operator using if statment
 - AND
 - OR
 - NOT

#### 1. AND
Returns True only if both conditions are true

```python
age = 25
income = 40000

if age > 18 and income > 30000:
    print("Eligible for loan")
```

#### 2. OR:
Returns True if at least one condition is true.


```python
city = "Chennai"

if city == "Chennai" or city == "Bangalore":
    print("City is in South India")
```

#### 3. NOT :
Reverses the truth value of a condition.




```python
is_logged_in = False

if not is_logged_in:
    print("Please log in")
```

### nested if:

A nested if statement in Python is an if statement inside another if statement. This allows for multiple levels of decision-making based on different conditions. It's useful when you have to check for multiple conditions, one depending on the result of another.

```python
# 1.  Nested if for Checking Multiple Conditions (Age and Height)
age = 20
height = 170

if age >= 18:
    if height >= 160:
        print("You are an adult and tall enough for the ride.")
    else:
        print("You are an adult but not tall enough for the ride.")
else:
    print("You are not an adult.")
```

```python
#nested if structure
balance = float(input('Enter your bank balance:'))
withdraw = float(input('Enter the amount you want to withdraw: '))
if withdraw <= balance:
  if withdraw > 500:
    print('This transaction is approved')
  else:
    print('transaction requires manager approval')
else:
  print('Insufficient funds.')
```

### 1. Using if with a List

```python
fruits = ['apple', 'banana', 'cherry']

if 'banana' in fruits:
    print("Banana is in the list.")
else:
    print("Banana is not in the list.")
```

### 2. Using if with a Tuple

```python
colors = ('red', 'green', 'blue')

if 'green' in colors:
    print("Green is in the tuple.")
else:
    print("Green is not in the tuple.")
```

```python
### 3. Using if with a Dictionary

person = {
    'name': 'John',
    'age': 30,
    'city': 'New York'
}

if 'age' in person:
    print("The key 'age' is present in the dictionary.")
else:
    print("The key 'age' is not present in the dictionary.")
```
